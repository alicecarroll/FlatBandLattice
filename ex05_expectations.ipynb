```python
# Given some model with fixed parameters, compute the hartree and cooper expectation values

# Plot <c^t c> and <c c> as a function of mu or delta, for different models.

import numpy as np
import matplotlib.pyplot as plt

from importlib import reload
from tblib import models, plotting
reload(models)
reload(plotting)

# Define model size
N = 4

model_list = [
    models.DSLmodel(N=N), 
    models.dDSLmodel(N=N), 
    models.LiebNmodel(N=N)
    ]

# Define Figure
fig, axes = plt.subplots(2, len(model_list),figsize=(4*len(model_list), 8))

nk = 2

mvals = np.linspace(0,1)
dvals = np.linspace(0,1)

for i, model in enumerate(model_list):
    axes[0, i] = plotting.plot_mean_fields(model, mvals, pname='mu', s=(N,N), nk=nk, ax=axes[0, i]);
    axes[1, i] = plotting.plot_mean_fields(model, dvals, pname='delta', s=(N,N), nk=nk, ax=axes[1, i]);
```
