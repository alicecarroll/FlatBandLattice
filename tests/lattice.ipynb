```python
import numpy as np
import matplotlib.pyplot as plt

import sys
sys.path.insert(0, "../") 

from importlib import reload
from tblib import lattice
reload(lattice)
 
fig, axes = plt.subplots(3, 4, figsize=(12,9))

lattices = [(
    lattice.DiagonallyStripedLattice(N=i), 
    lattice.dDiagonallyStripedLattice(N=i),
    lattice.LiebNLattice(N=i),
    ) for i in range(2,6)]

labels = [(
    f'DSL (N={i})', 
    f'dDSL (N={i})',
    f'Lieb-N (N={i})',
    ) for i in range(2,6)]

for i, lats in enumerate(lattices):
    for j, lat in enumerate(lats):
        if j!=2:
            field = lat.striped_props(props=np.arange(0,lat.N)/2)
        else:
            field = np.random.rand(len(lat.sites))*2
        _ = lat.plot_nn(axes[j, i], field=field, cmap='autumn')
        axes[j,i].set_title(labels[i][j])
```
