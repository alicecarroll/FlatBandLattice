```python
import numpy as np
import matplotlib.pyplot as plt

from importlib import reload
from tblib import hamiltonian, plotting
reload(hamiltonian)
reload(plotting)

# Define model size
N = 4

# Create model instances
dvals = [0.1, 0.5,]

# Define Figure
fig, axes = plt.subplots(len(dvals),6,figsize=(12,4*len(dvals)))

nk_dos = 30

for i, dval in enumerate(dvals):

    model = hamiltonian.DSLmodel(N=N)
    delta_vec = dval * np.ones(model.n)
    model.delta = delta_vec

    HBdG = model.get_HBdG()
    axes[i, 0] = plotting.plot_bands(HBdG, s=(N,N), nk=500, hsp_path='GYMXGMH', ax=axes[i, 0]);
    axes[i, 1] = plotting.plot_DOS(HBdG, nk=nk_dos, s=(N,N), elim=axes[i, 0].get_ylim(), ax=axes[i, 1], ylabel=None);


    model = hamiltonian.dDSLmodel(N=N)
    delta_vec = dval * np.ones(model.n)
    model.delta = delta_vec

    HBdG = model.get_HBdG()
    axes[i, 2] = plotting.plot_bands(HBdG, s=(N,N), nk=500, hsp_path='GYMXGMH', ax=axes[i, 2], ylabel=None);
    axes[i, 3] = plotting.plot_DOS(HBdG, nk=nk_dos, s=(N,N), elim=axes[i, 2].get_ylim(), ax=axes[i, 3], ylabel=None);

    model = hamiltonian.LiebNmodel(N=N)
    model.delta = dval

    HBdG = model.get_HBdG()
    axes[i, 4] = plotting.plot_bands(HBdG, s=(N,N), nk=500, hsp_path='GYMXGMH', ax=axes[i, 4], ylabel=None);
    axes[i, 5] = plotting.plot_DOS(HBdG, nk=nk_dos, s=(N,N), elim=axes[i, 4].get_ylim(), ax=axes[i, 5], ylabel=None);
```
