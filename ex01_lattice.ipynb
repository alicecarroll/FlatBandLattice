```python
# Runtime: ~seconds

import numpy as np
import matplotlib.pyplot as plt

from importlib import reload
from tblib import lattice
reload(lattice)
 
fig, axes = plt.subplots(3, 4, figsize=(12,9))

lattices = [(
    lattice.DiagonallyStripedLattice(N=i), 
    lattice.dDiagonallyStripedLattice(N=i),
    lattice.LiebNLattice(N=i),
    ) for i in range(1,5)]

labels = [(
    f'DSL (N={i})', 
    f'dDSL (N={i})',
    f'Lieb-N (N={i})',
    ) for i in range(1,5)]

for i, lats in enumerate(lattices):
    for j, lat in enumerate(lats):
        _ = lat.plot_nn(axes[j, i])
        axes[j,i].set_title(labels[i][j])
```
