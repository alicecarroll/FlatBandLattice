```python
N=5
f1li = []
f3li = []
pflist = []
kx=np.pi+0.1
ky=np.pi
summe=0
karr = np.linspace(-np.pi,0,N)
gammaz = np.kron(np.diag([1,-1]), np.eye(int(ham.dim/2)))

        
H = ham.Hk(kx,ky)[0]
dHdmy = ham.Hk(kx,ky,o='x')[0]
dHdny = ham.Hk(kx,ky,o='x')[0]

M1 = np.matmul(dHdmy,gammaz)
M2 = np.matmul(gammaz,dHdny)

evals, evec = np.linalg.eigh(H)
Evec = evec.T 

nE = [ham.fermidirac(E,o=0) for E in evals]
dnE = [ham.fermidirac(E,o=1) for E in evals]
for k,i in enumerate(evals):
    for l,j in enumerate(evals):
        if np.abs(i-j)<1e-3 or k==l:
            pf = -dnE[l]
        else:
            pf = (nE[k]-nE[l])/(j-i)
        if pf==0:
            summe+=0
            s=0
            

        else:
            f1 = np.matmul(np.conjugate(Evec[k]),np.matmul(dHdmy,Evec[l]))
            f2 = np.matmul(np.conjugate(Evec[l]),np.matmul(dHdny,Evec[k]))
            f1li.append(f1*f2)


            f3 = np.matmul(np.conjugate(Evec[k]),np.matmul(M1,Evec[l]))
            f4 = np.matmul(np.conjugate(Evec[l]),np.matmul(M2,Evec[k]))
            f3li.append(f3*f4)


            s = pf*(f1*f2-f3*f4)
            summe +=s/N**2

            pflist.append(pf)

                    
```
